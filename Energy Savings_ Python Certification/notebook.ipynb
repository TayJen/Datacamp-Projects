## Energy saved from recycling
<p>Did you know that recycling saves energy by reducing or eliminating the need to make materials from scratch? For example, aluminum can manufacturers can skip the energy-costly process of producing aluminum from ore by cleaning and melting recycled cans. Aluminum is classified as a non-ferrous metal.</p>
<p>Singapore has an ambitious goal of becoming a zero-waste nation. The amount of waste disposed of in Singapore has increased seven-fold over the last 40 years. At this rate, Semakau Landfill, Singaporeâ€™s only landfill, will run out of space by 2035. Making matters worse, Singapore has limited land for building new incineration plants or landfills.</p>
<p>The government would like to motivate citizens by sharing the total energy that the combined recycling efforts have saved every year. They have asked you to help them.</p>
<p>You have been provided with three datasets. The data come from different teams, so the names of waste types may differ.</p>

```python
df_1 = df_1[df_1['year'] >= 2015]
```

```python
df_1_waste = pd.pivot_table(df_1, values='total_waste_recycled_tonne', \
              index='year', columns='waste_type')
df_1_waste.head()
```

```python
df_1_rate = pd.pivot_table(df_1, values='recycling_rate', \
              index='year', columns='waste_type')
df_1_rate.head()
```

```python
df_1_waste.columns
```

```python
prep_cols = [('Ash & Sludge', 'Ash and sludge'),
             ('C&D', 'Construction debris'),
             ('Non-ferrous metal', 'Non-ferrous metals'),
             ('Others (stones, ceramic, rubber, etc.)', 'Others (stones, ceramics & rubber etc)'),
             ('Others (stones, ceramic, rubber, etc.)', 'Others (stones, ceramics & rubber etc.)'),
             ('Plastic', 'Plastics')]
```

```python
for i in [df_1_waste, df_1_rate]:
    for s1, s2 in prep_cols:
        i.loc[:, s1] = i.loc[:, [s1, s2]].sum(axis=1)
        i.drop(s2, axis=1, inplace=True)
        
df_1_waste.head()
```

```python
imp_cols = ['Plastic', 'Glass', 'Ferrous metal', 'Non-ferrous metal']

df_1_rate = df_1_rate.loc[:, imp_cols]
```

```python
df_2_waste = pd.pivot_table(df_2, values="Total Recycled ('000 tonnes)", \
              index='Year', columns='Waste Type')
df_2_waste.head()
```

```python
imp_cols = ['Plastics', 'Glass', 'Ferrous Metal', 'Non-Ferrous Metal']
df_2_waste = df_2_waste.loc[:, imp_cols]

df_2_res = df_2_waste
df_2_res.head()
```

```python
for i, col in enumerate(imp_cols):
    df_2_res.loc[:, col] = df_2_res.loc[:, col].mul(int(df_3_app.iloc[i, 1].split()[0]))
    
df_2_res.head()
```

```python
df_2_res['Result'] = df_2_res.sum(axis=1)
df_2_res.head()
```

```python
df_2_res.drop(imp_cols, axis=1, inplace=True)
df_2_res.head()
```

```python
results = np.concatenate((df_1_res.values.reshape(3), df_2_res.values.reshape(2) * 1000))

print(results)
```

```python
annual_energy_savings = pd.DataFrame(data={'year': range(2015, 2020), 'total_energy_saved': results})
annual_energy_savings.set_index('year', inplace=True)
annual_energy_savings.total_energy_saved = annual_energy_savings.total_energy_saved.astype('int')
annual_energy_savings.head()
```
